```python
largura_tela = 500
altura_tela = 400

pygame.init()

tela = pygame.display.set_mode((largura_tela, altura_tela))
pygame.display.set_caption("Pong Game")
relogio = pygame.time.Clock()

sair = False

pos_x_bola = largura_tela/2
pos_y_bola = altura_tela/2
raio_bola = 10

# Velocidade da bola em pixels no eixo x e y
vel_x = 1
vel_y = 1

while not sair:
    tela.fill("black")
    relogio.tick(100)

    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            sair = True

    bola = pygame.Rect(pos_x_bola, pos_y_bola, 2*raio_bola, 2*raio_bola)

    # Adicionamos a velocidade para as posições do retangulo criado para a bola
    pos_x_bola += vel_x
    pos_y_bola += vel_y

    pygame.draw.circle(tela, (255, 0, 0), (bola.centerx, bola.centery), raio_bola)

    pygame.display.update()
```

Rodando podemos ver que a bola saiu da tela, então devemos fazer com que ela fica dentro da tela

```python
while not sair:
    tela.fill("black")
    relogio.tick(100)

    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            sair = True

    bola = pygame.Rect(pos_x_bola, pos_y_bola, 2*raio_bola, 2*raio_bola)

    pos_x_bola += vel_x
    pos_y_bola += vel_y

    # Colisão com a tela
    # Eixo X
    if bola.right > largura_tela: # Se a bola passou do lado direito da tela
        vel_x = -vel_x # Alteramos a direção da bola para ir ao contrário
    elif bola.left < 0: # Se a bola passou do lado esquerdo da tela
        vel_x = -vel_x

    # Eixo Y
    if bola.bottom > altura_tela: # Se a bola passou do lado de baixo da tela
        vel_y = -vel_y # Mesma coisa do anterior mas no eixo Y
    elif bola.top < 0: # Se a bola passou do lado de cima da tela
        vel_y = -vel_y

    pygame.draw.circle(tela, (255, 0, 0), (bola.centerx, bola.centery), raio_bola)

    pygame.display.update()
```

```python
import pygame

largura_tela = 500
altura_tela = 400

pygame.init()

tela = pygame.display.set_mode((largura_tela, altura_tela))
pygame.display.set_caption("Pong Game")
relogio = pygame.time.Clock()

sair = False

pos_x_bola = largura_tela/2
pos_y_bola = altura_tela/2
raio_bola = 10

# Primeiro setamos a largura e altura das raquetes
largura_raq = 30
altura_raq = 100

# Agora setamos as variaveis de posição da raquete
pos_x_raq1 = 0
pos_y_raq1 = altura_tela/2 # Fazer aparecer na metade da tela

vel_x = 1
vel_y = 1

while not sair:
    tela.fill("black")
    relogio.tick(100)

    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            sair = True

    bola = pygame.Rect(pos_x_bola, pos_y_bola, 2*raio_bola, 2*raio_bola)

    # Igual a lógica do circulo, criamos um retangulo para a raquete
    raquete_1 = pygame.Rect(pos_x_raq1, pos_y_raq1, largura_raq, altura_raq)

    pos_x_bola += vel_x
    pos_y_bola += vel_y

    if bola.right > largura_tela:
        vel_x = -vel_x

        pos_x_bola = largura_tela - 2*raio_bola
    elif bola.left < 0: 
        vel_x = -vel_x

        pos_x_bola = 0

    if bola.bottom > altura_tela:
        vel_y = -vel_y

        pos_y_bola = altura_tela - 2*raio_bola
    elif bola.top < 0:
        vel_y = -vel_y

        pos_y_bola = 0

    pygame.draw.circle(tela, (255, 0, 0), (bola.centerx, bola.centery), raio_bola)

    # A diferença do circulo e retnagulo está no metodo para adicionar eles, parametros: superficie, cor, posição x, posição y, largura, altura
    pygame.draw.rect(tela, (255, 255, 255), raquete_1) # Os últimos parametros estão dentro do objeto raquete_1

    pygame.display.update()
```

Agora faremos com que a raquete se mova com as teclas W e S, além de checar colisão com a tela

```python
while not sair:
    tela.fill("black")
    relogio.tick(100)

    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            sair = True

    bola = pygame.Rect(pos_x_bola, pos_y_bola, 2*raio_bola, 2*raio_bola)
    
    raquete_1 = pygame.Rect(pos_x_raq1, pos_y_raq1, largura_raq, altura_raq)

    pos_x_bola += vel_x
    pos_y_bola += vel_y

    # Pega as teclas que o jogador apertou durante o loop
    keys = pygame.key.get_pressed()

    # Teclas do jogador 1
    if keys[pygame.K_w]: # Se apertar W faremos com que ele suba
        pos_y_raq1 -= 1 
    elif keys[pygame.K_s]: # Se apertar S faremos com que ele desça
        pos_y_raq1 += 1

    # Colisão igual a do circulo
    if raquete_1.bottom > altura_tela:
        pos_y_raq1 = altura_tela - altura_raq
    elif raquete_1.top < 0:
        pos_y_raq1 = 0

    if bola.right > largura_tela:
        vel_x = -vel_x

        pos_x_bola = largura_tela - 2*raio_bola
    elif bola.left < 0: 
        vel_x = -vel_x

        pos_x_bola = 0

    if bola.bottom > altura_tela:
        vel_y = -vel_y

        pos_y_bola = altura_tela - 2*raio_bola
    elif bola.top < 0:
        vel_y = -vel_y

        pos_y_bola = 0

    pygame.draw.circle(tela, (255, 0, 0), (bola.centerx, bola.centery), raio_bola)

    pygame.draw.rect(tela, (255, 255, 255), raquete_1)

    pygame.display.update()
```

Agora iremos chechar a colisão da bola com a raquete, e verificar de qual direção foi

```python
while not sair:
    tela.fill("black")
    relogio.tick(100)

    for evento in pygame.event.get():
        if evento.type == pygame.QUIT:
            sair = True

    bola = pygame.Rect(pos_x_bola, pos_y_bola, 2*raio_bola, 2*raio_bola)
    
    raquete_1 = pygame.Rect(pos_x_raq1, pos_y_raq1, largura_raq, altura_raq)

    pos_x_bola += vel_x
    pos_y_bola += vel_y

    keys = pygame.key.get_pressed()

    if keys[pygame.K_w]:
        pos_y_raq1 -= 1 
    elif keys[pygame.K_s]:
        pos_y_raq1 += 1

    if raquete_1.bottom > altura_tela:
        pos_y_raq1 = altura_tela - altura_raq
    elif raquete_1.top < 0:
        pos_y_raq1 = 0

    if bola.right > largura_tela:
        vel_x = -vel_x

        pos_x_bola = largura_tela - 2*raio_bola
    elif bola.left < 0: 
        vel_x = -vel_x

        pos_x_bola = 0

    if bola.bottom > altura_tela:
        vel_y = -vel_y

        pos_y_bola = altura_tela - 2*raio_bola
    elif bola.top < 0:
        vel_y = -vel_y

        pos_y_bola = 0

    # Checa se houve colisão entre a raquete e a bola, não precisa de colisão pela esquerda nessa raquete
    if raquete_1.colliderect(bola):
        if raquete_1.right < bola.right: # Colisão pela esquerda ou direita
            vel_x = -vel_x

            pos_x_bola = raquete_1.right
        elif raquete_1.top > bola.top: # Colisão pelo topo
            vel_y = -vel_y

            pos_y_bola = raquete_1.top - 2*raio_bola # Tem que tirar o tamanho o diametro da bola se não buga muito
        else: # Colisão por baixo
            vel_y = -vel_y

            pos_y_bola = raquete_1.bottom
        
    pygame.draw.circle(tela, (255, 0, 0), (bola.centerx, bola.centery), raio_bola)

    pygame.draw.rect(tela, (255, 255, 255), raquete_1)

    pygame.display.update()
```
